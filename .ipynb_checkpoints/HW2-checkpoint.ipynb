```python
def lemmatize(sent):
    normalized_words = []
    pymrth_analyzer = MorphAnalyzer()
    for i in range(len(sent)):
        word = pymrth_analyzer.parse(sent[i])
        normal_word = word[0].normal_form
        normalized_words.append(normal_word)
        
    return normalized_words
```

```python
def tokenize(file):
    f = open(file, 'r', encoding='utf-8')
    for line in f:
        for ch in string.punctuation:
            line = line.replace(ch,"")
        sent = word_tokenize(line)
        filtered_sent = [word for word in sent if word not in noise]
        lemmas = lemmatize(filtered_sent)
        return lemmas
        #print(sent)
```

```python
'''TASK#3
    инвертированный список терминов 
    Создаем словарь,
    dict1 содержит слова,
    dict2 содержит номера документов соответсвующих слов
    {word:[set_of_all_doc_ids]}
'''
def inverted_term_list(tokens):
    dict1 = {}
    dict2 = {}
    for i in range(0, len(tokens)):
        for token in tokens[i]:
            #print(i, token)
            if(token == ''): continue
            if token.lower() not in dict1:
                dict1[token.lower()] = set()         #новое слово / количество слов документе
                dict2[token.lower()] = 1
            else:
                dict2[token.lower()] += 1
            dict1[token.lower()].add(i)            #добавляем номер документа
    print(dict1)
    print("======")
    print(dict2)
    return dict1, dict2
```

```python
term_dict, word_count = inverted_term_list(tokens)
```

```python
def search(query):
    import re

    result_set=set() # итоговое множество
    operation=None
    print(re.split("\s*(\&|\|)\s*", query))
    words = re.split("\s*(\&|\|)\s*", query)
#    global_set = set()
#     for word in words:
#         if word not in ['&','|']:
#             if word[0] == '!':
#                 realword=word[1:]
#             else:
#                 realword=word
#             global_set+=set(term_dict.get(realword))
#             print(realword)
    firstInverted = None #запоминаем на всякий случай, если & следующий оператор, чтобы убрать из множества
    for idx, word in enumerate(words, start=0):
        print(idx, word)

        inverted=False # для "!"

        if word in ['&','|']:
            operation=word
            continue

        if word[0] == '!':
            if operation == '|':
                continue
            # идет операция AND, поэтому надо убрать множество
            if idx==0:
                firstInverted = word[1:]
            inverted = True
            # the word is inverted!
            realword = word[1:]
            print(idx, realword)
        else:
            realword=word

        if operation is not None:
            # now we need to match the key and the filenames it contains:
            current_set=set(term_dict.get(realword))
            print(idx, current_set)

            if operation == '&':
                if inverted == True:
                    print(idx, "excluding", current_set, " from ", result_set)
                    result_set -= current_set
                    print("res ", result_set) 
                elif result_set==set():
                    print(idx, "|", current_set, " with ", result_set)
                    result_set |= current_set
                    print("res ", result_set)
                else:
                    print(idx, "&&", current_set, " with ", result_set)
                    result_set &= current_set
                    print("res ", result_set)
            elif operation == '|':
                result_set |= current_set
        elif operation is None and inverted == False: #для 1|....
            current_set=set(term_dict.get(realword))
            print(idx, current_set)
            result_set |= current_set
        
        if idx==2 and firstInverted is not None: #для !1 & ...
            current_set=set(term_dict.get(firstInverted))
            print(idx, "excluding", current_set, " from ", result_set)
            result_set -= current_set
            print(idx, "res ", result_set)
        operation=None
    if (result_set==set()):
        print("Not found!")
    else:
        print(result_set)
```

```python
import collections
def compute_tf(term_dict, tokenList):#отношение числа вхождений некоторого слова к общему числу слов документа
    dict = {}
    #На вход берем текст в виде списка (list) слов
    #Считаем частотность всех терминов во входном массиве с помощью 
    #метода Counter библиотеки collections
    tf_text = collections.Counter(tokenList)
    for term in term_dict:
        #для каждого слова в tf_text считаем TF путём деления
        #встречаемости слова на общее количество слов в тексте
        dict[term] = tf_text.get(term)/float(len(tokenList))
    #возвращаем объект типа Counter c TF всех слов текста
    return dict
```
